```python
num_plots = len(catcols)
num_rows = (num_plots + 1) // 2  # Add 1 and divide by 2 to round up for odd numbers
num_cols = 2
# Loop through each feature and create a countplot
for i, feature in enumerate(catcols, 1):
    plt.subplot(num_rows, num_cols, i)
    sns.countplot(x=feature, data=df, palette='Wistia')
    plt.title(f'Bar Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)

# Adjust layout to prevent overlap of subplots
plt.tight_layout()
plt.show()
```

```python
high_cor_cols =['emp.var.rate','euribor3m','nr.employed']
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("bank-additional.csv")


df1 = df.copy()
df1.columns
```

```python
df1.drop(high_cor_cols,inplace=True,axis=1)
df1.columns
```

```python
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_encoded=df1.apply(le.fit_transform)
df_encoded
```

```python
df_encoded['y'].value_counts()
```

```python
x = df_encoded.drop('y',axis=1)
y = df_encoded['y']
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))
def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)
    print('Testing Score',test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(criterion='entropy',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
ypred_dt=dt.predict(x_test)
print(ypred_dt)
```

```python
eval_model(y_test,ypred_dt)
```
